## 1. Introduction (ভূমিকা)

**Long Short-Term Memory (LSTM)** হলো Recurrent Neural Network (RNN)-এর একটি বিশেষ উন্নত আর্কিটেকচার, যা ১৯৯৭ সালে **Sepp Hochreiter** এবং **Jürgen Schmidhuber** তৈরি করেন।

সাধারণ RNN যেখানে দীর্ঘ sequence-এ তথ্য "ভুলে যায়", সেখানে LSTM দুইটি আলাদা memory রেখে — **Long-Term Memory** ও **Short-Term Memory** — সেই সমস্যা সমাধান করে।

> 💡 **গুরুত্বপূর্ণ কথা:** আজকের যুগের GPT, BERT, Transformer, ChatGPT — এই সমস্ত আধুনিক AI-এর foundational step হলো RNN → LSTM। তাই LSTM না বুঝলে LLM বোঝা সম্ভব নয়।
