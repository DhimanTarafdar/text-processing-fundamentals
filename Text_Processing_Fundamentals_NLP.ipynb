## 🔹 NLP (Natural Language Processing) কী?

**NLP (Natural Language Processing)** হলো **Artificial Intelligence (AI)** এর একটি শাখা, যার মাধ্যমে কম্পিউটারকে মানুষের ভাষা (যেমন বাংলা, ইংরেজি) **পড়া, বোঝা, বিশ্লেষণ করা এবং তৈরি করা (generate)** শেখানো হয়। সহজ ভাষায় বললে, NLP হলো এমন একটি টেকনিক যার মাধ্যমে কম্পিউটার **text** বা **speech** থেকে human language প্রসেস করতে পারে।

**ব্যবহারিক ক্ষেত্র (Use Cases):**
- Chatbot ও Virtual Assistant (Siri, Alexa, ChatGPT)
- Machine Translation (Google Translate)
- Sentiment Analysis (কোনো রিভিউ Positive না Negative তা বোঝা)
- Spam Email Detection
- Text Summarization
- Speech Recognition (Voice to Text)
