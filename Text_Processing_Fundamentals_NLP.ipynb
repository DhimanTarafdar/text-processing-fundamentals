## 🔹 NLP (Natural Language Processing) কী?

**NLP (Natural Language Processing)** হলো **Artificial Intelligence (AI)** এর একটি শাখা, যার মাধ্যমে কম্পিউটারকে মানুষের ভাষা (যেমন বাংলা, ইংরেজি) **পড়া, বোঝা, বিশ্লেষণ করা এবং তৈরি করা (generate)** শেখানো হয়। সহজ ভাষায় বললে, NLP হলো এমন একটি টেকনিক যার মাধ্যমে কম্পিউটার **text** বা **speech** থেকে human language প্রসেস করতে পারে।

**ব্যবহারিক ক্ষেত্র (Use Cases):**
- Chatbot ও Virtual Assistant (Siri, Alexa, ChatGPT)
- Machine Translation (Google Translate)
- Sentiment Analysis (কোনো রিভিউ Positive না Negative তা বোঝা)
- Spam Email Detection
- Text Summarization
- Speech Recognition (Voice to Text)

## 🔹 NLP Basic Pipeline (ধাপে ধাপে)

কোনো **raw text** কে কাজে লাগানোর মতো (useful) তথ্যে রূপান্তর করতে নিচের ধাপগুলো অনুসরণ করা হয় —

1. **Text Acquisition (Data Collection)** — raw text data সংগ্রহ করা (website, social media, document ইত্যাদি থেকে)।
2. **Text Cleaning / Preprocessing** — অপ্রয়োজনীয় characters, HTML tag, punctuation, extra space বাদ দেওয়া। 👉 **এই স্টেপেই Regular Expression সবচেয়ে বেশি ব্যবহৃত হয়।**
3. **Tokenization** — sentence কে word বা sentence এ ভাগ করা।
4. **Stop word Removal** — "is, the, a" এর মতো কম গুরুত্বপূর্ণ শব্দ বাদ দেওয়া।
5. **Stemming / Lemmatization** — শব্দকে মূল রূপে (root form) আনা (যেমন: running → run)।
6. **Feature Extraction** — text কে numeric vector এ রূপান্তর করা (Bag of Words, TF-IDF, Word Embedding)।
7. **Model Building** — Machine Learning / Deep Learning model train করা।
8. **Evaluation** — model এর performance যাচাই করা।
> 🔑 এই পাইপলাইনের ২নং স্টেপ (Cleaning/Preprocessing) -এ raw text থেকে digit, email, special character, URL ইত্যাদি খুঁজে বের করতে বা remove করতে **Regular Expression** একটি অপরিহার্য টুল।

## 🔹 Regular Expression (Regex) কী?

**Regular Expression (Regex)** হলো এক ধরনের **pattern / sequence of characters**, যা কোনো text এর মধ্যে নির্দিষ্ট pattern **খুঁজে বের করতে (search)**, **মিলিয়ে দেখতে (match)**, **পরিবর্তন করতে (replace)** বা **ভাগ করতে (split)** ব্যবহার করা হয়।

Python এ regex ব্যবহারের জন্য built-in **`re`** module ব্যবহার করা হয়।

**Regex এর প্রয়োজনীয়তা (Use Cases):**
- Text থেকে phone number, email address, date খুঁজে বের করা
- Data Validation (যেমন: email বা password এর format সঠিক কিনা চেক করা)
- Text Cleaning (অপ্রয়োজনীয় character বাদ দেওয়া)
- Find & Replace করা
- নির্দিষ্ট delimiter দিয়ে text কে ভাগ (split) করা
---

````python
import re

txt = "The rain spain"
x = re.findall(r"[a-b]", txt)
print(x)
````

## 1️⃣ Character Set `[ ]`

**Use case:** Character set `[ ]` দিয়ে কোনো নির্দিষ্ট set/range এর মধ্যে থাকা character match করা হয়। যেমন `[a-b]` মানে হলো 'a' অথবা 'b' এর সাথে match করবে।

```python
import re

txt = "The rain spain"
x = re.findall(r"[a-b]", txt)
print(x)
```

**কোড ব্যাখ্যা:**
- `import re` → Python এর built-in regex module import করা হলো।
- `r"[a-b]"` pattern টি text এর ভেতরে শুধু **'a'** অথবা **'b'** (case-sensitive) character গুলো খুঁজবে।
- `re.findall()` ফাংশন pattern এর সাথে মিলে যাওয়া সব character কে একটি **list** এ return করে।
**Output:** `['a', 'a']`  (শব্দ "rain" ও "spain" এ একটি করে 'a' আছে; 'b' কোথাও নেই)
